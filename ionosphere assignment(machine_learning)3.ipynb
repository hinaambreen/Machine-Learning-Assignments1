```python
#import libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from keras import models,layers
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#Load Data
x = pd.read_csv("C:\\Users\\hina\\Downloads\\ML_Datasets-main\\ionosphere_data.csv")
```

```python
#Check Missing Values
x.isnull().any()
x.isnull().sum()
```

```python
#Shuffle the data if needed.
from sklearn.utils import shuffle
x = shuffle(x)
```

```python
#Encode labels.
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
x['label'] = labelencoder.fit_transform(x['label'])
x.label.head()
```

```python
#Split x data into features and labels
features = x.drop("label",axis=1)
labels = x["label"]
labels.head()
#labels.value_counts() 
#features.head()
```

```python
#Split into 60 and 40 ratio.
train_features,test_features,train_label,test_label = train_test_split(features,labels,test_size=0.4,random_state=0)
```

```python
#Standardized the Input Variables.
sc = StandardScaler()
train_features = sc.fit_transform(train_features)
test_features = sc.fit_transform(test_features)
print(test_features.shape)
print(test_label.shape)
print(train_features.shape)
print(train_label.shape)
```

```python
#Model : 1 hidden layers including 16 unit.
#Compilation Step (Note : Its a Binary problem , select loss , metrics according to it)
def build_model():
    model = models.Sequential()
    model.add(layers.Dense(16,activation="relu",input_shape= (train_features.shape[1],)))
    model.add(layers.Dense(16,activation="relu"))
    model.add(layers.Dense(1,activation='sigmoid'))
    model.compile(optimizer="rmsprop",loss = "binary_crossentropy",metrics = ["accuracy"])
    return model
```

```python
#Train the Model with Epochs (100).
model = build_model()
history = model.fit(train_features,train_label,batch_size=2, epochs=100)
```

```python
#Prediction should be > 92%
#Evaluation Step
loss,acc = model.evaluate(test_features,test_label)
```

```python
#Prediction
pred = model.predict(test_features)
pred[:10]
```
